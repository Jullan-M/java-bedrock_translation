```python
from comparative_translator import Locale_Translation
from utilities import dict2json, file2dict
from tqdm import tqdm
from excel_tools import Ranked_Translations
```

```python
en_new = file2dict("Locales_1_21/en_US_1_21.lang")
en_old = file2dict("Locales_1_7/en_US_1_7.lang")
sme_old = file2dict("Locales_1_7/se_NO.lang")
```

```python
en_new_keys = set(en_new.keys())
en_old_keys = set(en_old.keys())
sme_old_keys = set(sme_old.keys())

shared_keys = sme_old_keys.intersection(en_new_keys).intersection(en_old_keys)
print(len(shared_keys))
```

```python
changed = {}
for k in shared_keys:
    if en_old[k].lower() != en_new[k].lower():
        changed[k] = [en_old[k] , en_new[k]]
        print(k)
dict2json(changed, "changed-1_7-1_21.json")
print(len(changed))
```

```python
ranked = file2dict("ranked.json")

for k in changed.keys():
    ranked[k] = {
        "value": ranked[k]["value"],
        "rank": "FROM_OLD_LANG",
        "quality": 0
    }
dict2json(ranked, "ranked.json")

rt = Ranked_Translations(ranked)
rt.to_excel("ranked_sme_from_nob1.xlsx", "Locales_1_21/en_US_1_21.lang")
```
